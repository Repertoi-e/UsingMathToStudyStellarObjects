## Обобщение
Астрономите използват математика през цялото време. Въпреки, че физиката дава обяснение на разни феномени, понякога тя се оказва недостатъчна и затова картинката се допълва чрез математически модели.

Въпреки, че в днешно време физиката е напреднала значително и вече може да даде обяснение на почти всяко движение на обектите в нашата Слънчева система (а и извън нея), нека разгледаме как можем само чрез математика и наблюдение да изчислим най-базовото: колко далеч са небесните обекти едни от други и колко големи са те.

След това ще разгледаме как се описват техните движения и орбити.

Накрая ще видим какво точно е денонощие и как да изчислим уравнението на времето - функция, която ни показва колко се отдалечава часовникът ни от действителната продължителност на денонощието. 

## Дефиниции

Преди да започнем, нека разгледаме няколко неща, които ще ни трябват по-нататък.

### Паралакс
Паралакс е видимото изместване на един обект, ако той е наблюдаван от две различни точки. Например, ако гледаме един обект първо само с лявото си око, а после само с дясното, ще забележим, че в единия случай той е по-надясно, а в другия – по-наляво. Знаейки разстоянието между очите си и ъгъла между двата образа, можем да пресметнем разстоянието между нас и обекта. 

### Ексцентрицитет на елипса
Ексцентрицитет на елипса е отношението на фокусното разстояние $F_1F_2$ и $C$, където $C = F_1M + F_2M$, а $M$ е дадена точка от елипсата. Ексцентрицитетът характеризира разтеглеността на елипсата – колкото е по-близък до 0, толкова повече елипсата наподобява окръжност, и обратното – колкото е по-близък до 1, толкова тя е по-издължена.
Ексцентрицитет на орбитата на земята е около $0.0167$.

## Определяне на разстоянието между Земята и друго небесно тяло

За определяне на разстоянието между Земята и някое друго небесно тяло извън Слънчевата система (звезда или планета) най-често се използва явлението паралакс. 

Същото можем да направим и с небесно тяло. Нека Земята е в положение $ Е_1 $, а ние се намираме в точка $ А $, Слънцето е в точка $ S $, небесното тяло, до което искаме да намерим разстоянието е в точка $ О $ и Земята се върти обратно на часовниковата стрелка. Това е илюстрирано на фиг. 1 (мащабите не са спазени).

<center><img src="./Math Project Images/fig1.jpg" width=500px/></center>

Наблюдаваме небесното тяло по изгрев, като гледаме право напред спрямо Вселената.
Правата $а$ е линията на погледа ни. Тогава виждаме $О$ под ъгъл $\alpha$ и $О$ е на запад от Слънцето, което изгрява от изток.
След шест месеца, когато Земята се намира в $Е_2$, правим същото наблюдение, но по залез. Правата $b$ ще е линията на погледа ни, а ъгълът, под който виждаме $О$, ще е $\alpha'$. Ъгълът между двата образа може да бъде измерен и ще е равен на $\alpha + \alpha'$.

$Е_2$ е положението на Земята шест месеца след $E_1 \implies \rm Е_2Е_1$ е диаметър на орбитата на Земята.
Позициите $Е_1$ и $Е_2$ са избрани, така че отсечката $\rm SO$, която свързва Слънцето и небесното тяло, да е перпендикулярна на диаметъра, 
$$\implies \rm SO \verb| е височина и медиана в |\vartriangle \rm Е_1ОЕ_2$$ 
$$\implies \vartriangle \rm Е_1ОЕ_2 \verb| е равнобедрен|$$ 
$$\implies \angle \rm ОЕ_1Е_2 = \angle \rmОЕ_2Е_1$$

Правите $а$ и $b$ са перпендикулярни на $\rm Е_1Е_2$ 
$$\implies \angle \rm ОЕ_1Е_2 = 90^{\circ} - \alpha'$$
$$\angle \rm ОЕ_2Е_1 = 90^{\circ} - \alpha' \implies \alpha = \alpha'$$
$$\implies \verb|ъгълът между двата образа е равен на | 2\alpha$$ 
$$\implies \verb|вече можем да намерим | \alpha$$

$$\vartriangle \rm Е_1ОS \verb| е правоъгълен | (\rm SO \perp \rm Е_2Е_1) \implies \cos(90^{\circ} - \alpha) = \frac{\rm SE_1}{\rm{ОЕ_1}}$$
$$\implies \rm ОЕ_1 = \frac{\rm{SE_1}}{\sin\alpha}$$

$\rm {SE_1}$ e разстоянието между Слънцето и Земята. То  вече е известно и в астрономията се означава с $1$ AU (1 астрономическа единица).
$$ \implies \rm{ОЕ_1} = \frac{1}{\sin\alpha} $$
Ако търсим разстоянието между Слънцето и небесното тяло $О$, то ще е равно на $ctg\ \alpha$.
Ако искаме да намерим разстоянието до друго небесно тяло (например $Т$), трябва да изберем други подходящи положения $Е_1$ и $Е_2$, от които да го наблюдаваме.

## Определяне на размера на небесно тяло

След като сме определили разстоянието между Земята и дадено небесно тяло, можем да намерим и неговия размер. За целта ще използваме **ъгловия размер** на тялото, което наблюдаваме. 

Ъгловият  размер е ъгълът, под който виждаме тяло, разположено на дадено разстояние от нас. Този ъгъл може да бъде измерен.
(Напр. известно е че небесни тела, които имат размера на кутрето ни на нощното небе имат ъглов размер $1^{\circ}$.)

<center><img src="./Math Project Images/fig2.jpg" width=250px/></center>

Нека гледаме небесно тяло с център $О$, както човечето от фиг. 2 (мащабите не са спазени). Виждаме небесното тяло под ъгъл $\alpha$. Размерът на тялото ще бъде отсечката $\rm АB$. Ако си представим, че ние сме в центъра на окръжност $k(C, R)$ с радиус разстоянието между Земята и небесното тяло, тогава отношението на дължината на дъгата $\rm АВ$ (нека я означим с $x$) към дължината на окръжността ще е равно на отношението на $\alpha$ към $360^{\circ}$.
$$\implies x = 2\pi R \frac{\alpha}{360^{\circ}}.$$
Тъй като $\alpha$ е изключително малък ъгъл, можем да направим приближението, че отсечката $\rm АВ$ и дъгата $\rm \widehat{АВ}$ съвпадат. Тогава $x$ ще е диаметър на небесното тяло. Ако приемем, че небесното тяло е идеална сфера, можем да намерим обема $\rm{V} = \frac{1}{6}x^3\pi$ и повърхнината му $\rm{S} = x^2\pi$.

## Определяне орбитата на небесно тяло

Всяка частица с маса привлича другите в близост (гравитацията), като колкото по-голяма е масата, толкова по-силно им действа. След Големия взрив всички частици са се движели хаотично и постепенно са започнали да се групират, като всяка група се завърта около общ център. Затова небесните тела се въртят около собствената си ос и около друго по-голямо тяло, което ги привлича.

Пътят, който те изминават, се нарича **орбита** и той се определя единствено от два параметъра – скоростта, с която се движат, и гравитационната сила, която им действа. Ако орбитата е окръжност, гравитационната сила, която се стреми да привлече тялото към центъра на системата, и центробежната сила, която действа в обратна посока, се уравновесяват.

$$ F = \frac{GMm}{r^2} = \frac{v^2 m}{r} \implies v = \sqrt{\frac{GM}{r}} $$

$G = 6.67408 × 10^{-11} m^3 kg^{-1} s^{-2}$ e гравитационната константа, $M$ и $m$ са масите съответно на централното и орбитиращото тяло, $r$ – радиусът на орбитата, $v$ е скорост на движение на орбитиращото тяло. 

Интересното е, че скоростта не зависи от масата на орбитиращото тяло.

### Тяло в кръгова орбита около земята

С горната изведена формула можем да пресменем с каква скорост се движи тяло, в кръгова орбита около Земята - напр. сателит:

Масата на земята е $M = 5.972 × 10^{24} kg$, радиусът й e $r_e = 6371\ km \implies r = r_e + h$, където $h$ е височината (в метри), на която се намира сателита.

$$ v = \sqrt{\frac{GM}{r_e + h}} = \sqrt{\frac{6.67408 × 10^{-11} × 5.972 × 10^{24} m^3 s^{-2}}{h + 6371000\ m}} = \frac{19964400}{\sqrt{h + 6371000}} \frac{m}{s}$$

### Закони на Кеплер

#### Първи закон
Орбитите на планетите са елипси, в единия от фокусите, на които се намира Слънцето. Те са много слабо сплеснати и почти не се отличават от окръжност.

Поради външните въздействия орбитата не може да бъде точна окръжност. Например, ако друго тяло (астероид, комета) случайно удари планетата обикаляща по кръгова орбита, то ще наруши равновесието между гравитационната и центробежната сила и разстоянието между планетата и звездата ще се увеличи. 

<center><img src="./Math Project Images/kepler-1.png"/ width=400px></center>

#### Втори закон

При движението на планетите около Слънцето техните радиус-вектори описват равни площи за равни интервали от време. От тук следва следва, че когато планетите преминават през **перихелия** си се движат с максимална скорост, а когато преминават през **афелия** – с минимална. Земята преминава през своя перихелий в началото на януари, а през афелия си – през юли.

<center><img src="./Math Project Images/kepler-2.png"/ width=400px></center>

#### Трети закон

Квадратите на периодите на обикаляне $T$ на планетите около Слънцето се отнасят, както кубовете на големите им полуоси $a$ до Слънцето:

<center><img src="./Math Project Images/kepler-3.jpg"/ width=500px></center>

Може да се докаже като видим, че гравитационната сила е равна на масата на тялото $m$, умножена по ъгловото му ускорение $a_c$, което от своя страна е равно на квадрата на скоростта $v$ на тялото в орбита към радиуса на орбитата $r$.

$$ F = ma_c = m\frac{v^2}{r} $$
    
$$ G\frac{mM}{r^2} = \frac{mv^2}{r} $$
$$ v^2 = G\frac{M}{r} $$

От $T$ - периодът на орбитата:
$$ v = \frac{2\pi r}{T} $$

$$ \implies G\frac{M}{r} = \frac{4\pi^2 r^2}{T^2} $$

$$ \implies T^2 = \frac{4\pi^2 r^3}{GM} $$

Така може да се изрази и масата на централното тяло:

$$ M = \frac{4\pi^2 r^3}{GT^2} $$ 

Или да получим третия закон на Кеплер:

$$ \frac{T_1^2}{T_2^2} = \frac{a_1^3}{a_2^3} $$
$$ \implies \frac{T_1^2}{a_1^3} = \frac{T_2^2}{a_2^3} = \frac{4\pi^2}{GM} = const$$

### Друби видове орбити

Телата в космоса могат да се движат и по парабола. Това се случва, когато биват привлечени от тяло и влизат в орбита около него или когато излизат от обхвата му (изстрелване на космически кораб извън орбита около Земята). В този случай минималната нужна скорост е с $\sqrt{2}$ по-голяма от минималната скорост в орбита.

Друга възможна траектория е хиперболичната. В този скучай тяло, свободно летящо из Космоса, бива привлечено от друго по-голямо, с което се променя посоката му на движение, но то не се задържа в орбита.

Обекти на параболични или хиперболични траектории не се считат за част от планетарната система.

### Инклинация на орбита

Инклинацията е наклон на орбитата на дадено небесно тяло. Тя представлява ъгълът между равнината на орбитата на даденото небесно тяло и изходната равнина (обикновено екватора на централното тяло или равнината на орбитата на Земята) в градуси.

<center><img src="./Math Project Images/inclination.png" width=400px/></center>

Нека равнината $\alpha$ е определена от екватора на централното тяло и изберем координатна система с център т. $О$. $ О_z \perp \alpha$.
Ако $\beta$ е равнината, определена от орбитата на второто тяло и $ \rm{EC} \perp \beta$, нека $\rm{OD} = \rm{EC} = h$.
В избраната координатна система z-компонентата на $h$ е $\rm{OF}$. Можем да определим ъгъла на инклинация ($i$) от $\vartriangle \rm{ODF}$:
$$ \angle \rm{ODF} = 90^{\circ} \implies \cos i = \frac{h_z}{|h|}$$

## Определяне на денонощието на дадена планета

Денонощието е мерна единица приблизително равна на времето, което отнема на Земята да се завърти веднъж около оста си.
Ние си представяме денонощието да започва и свършва в 12:00 на обяд, т.е. когато Слънцето е най-високо. 
Всъщност Земята освен около оста си се върти около Слънцето, което означава, че трябва да се завърти още малко (общо ~$361^{\circ}$) преди да сочи пак право към слънцето.

Едно **звездно денонощие** продължава 23 часа 56 минути 4 секунди (завъртане спрямо звездите, т.е само около оста си), а едно **слънчево денонощие** (завъртане спрямо Слънцето) - *средно* около 24 часа, с което сме свикнали.

<center><img src="./Math Project Images/solar-day.gif"/></center>

Но 24 часа е *средната* продължителност на слънчевото денонощие. Тя се променя през годината. Това се случва поради две причини:
* Земята няма точно кръгла орбита, когато е в своя афелий (юли), скоростта на Земята е най-малка, а когато е в своя перихелий (януари) е най-голяма.

<center><img src="./Math Project Images/orbit-complications.gif"/></center>

* Оста на въртене на Земята е наклонена с около $23.4^{\circ}$. В началото на лятото и зимата, Слънцето се движи почти успоредно на небесния екватор и земната ос е наклонена към Слънцето, а орбиталното движение на Слънцето е насочено почти точно от изток на запад - скоростта на изменение на часовия ъгъл на Слънцето е минимална. В началото на пролетта и есента, оста на Земята е изместена към Слънцето. Скоростта на изменение на часовия ъгъл на Слънцето е максимална.

Ако отчетем тези два факта, можем да изведем функция, която описва продължителността на слънчевото денонощие през всеки ден от годината.

Нищо тук не е конкретно за Земята. Всъщност всяка планета си има "година" и "денонощие" и обикновенно те се измерват спрямо земните. Ако знаем за колко време обикаля една планета около звездата си, знаем ексцентрицитета на нейната орбита, както и наклона на оста й, може да изчислим подобна функция за продължителността на денонощието на която и да е планета. В идеалния случай, когато ексцентрицитета е 0, т.е. орбитата е окръжност, а оста й е перпендикулярна на равнината на орбитата, т.е. без наклон, денонощието е едно и също за всеки "ден" от годината на планетата.

Както видяхме обаче Земята няма точно кръгла орбита, а оста й е под сравнително голям ъгъл. Вместо часовниците ни да измерват времето по-бързо или по-бавно в зависимост от позицията на Земята всъщност сме взели средната продължителност - 24 часа.

Можем да видим колко "грешат" часовниците като намерим т.нар. уравнение на времето.

### Уравнение на времето

По дефиниция уравнението на времето е разликата между средното слънчево време (ССВ) и истинското слънчево време (ИСВ).

$$\rm УВ = ССВ\ - \ ИСВ$$

Нека $W$ е средната орбитална скорост на Земята измерена в градуси за 1 ден.

$$ W = \frac{360^{\circ}}{365.24 \verb| дни|} $$

$A$ e ъгълът, който Земята изминава по орбитата си със своята средна скорост от зимното слънцестоене до даден ден през годината. $D$ е броят дни от 1-ви януари, a 10 дни е разликата между зимното слънцестоене (21-ви декември) и 1-ви януари.

$$ A = W(D+10)$$

$B$ e равен на $A$, но с компенсация за **ексцентрицитета** на орбитата на земята ($0.0167$). Числото 2 са дните от 1-ви януари до земния перихелий (деня в който орбиталната скорост е максимална).

$$ B = A + 0.0167\frac{360^{\circ}}{\pi}\sin(W(D-2))$$

$C$ е разликата между ъгъла на движение със средна скорост и точната скорост спрямо равнината на екватора. Числото $23.44$ е именно наклона на земната ос.

$$ C = \frac{A - \arctan\frac{\tan B}{\cos 23.44^{\circ}}}{180^{\circ}} $$

Делим на $180^{\circ}$, за да получим разликата в полузавъртания ($\pi$ радиани). Понеже функцията $\arctan$ (обратната функция на тангенс - ако знаем стойността на тангенс, тя ни казва ъгъла) има няколко стойности за даден $x$ - именно заради периодичността на $\tan x$. Тогава, за да получим правилната стойност, взимаме само стойността на $C$ след десетичната запетая. Тъй като сме разделили на $180^{\circ}$, частта от $C$ преди запетаята представлява колко допълнителни полузавъртания сме изчислили, които са дошли от периодичността на $\tan x$.

$$ \implies УВ = 720(C - \lceil C \rfloor)$$

Символът $\lceil x \rfloor$ означава закръгляне до най-близкото цяло число. Следователно $C - \lceil C \rfloor$ е равно на стойността на $C$ само след десетичната запетая. 

Умножаваме по 720 - броят на минутите (12 часа), за които земята прави полузавъртане спрямо Слънцето.

Ако напишем формулите в някой калкулатор получаваме графика с колко минути часовникът ни изостава за всеки ден от годината. Когато функцията пресича $O_x$, тогава денонощието трае точно 24 часа. Това се случва на 4 дни в годината: 15 април, 13 юни, 1 септември и 25 декември. Функцията е с период 365.

Тези графики отдолу са на двата отделни фактора (орбиталния ексцентрицитет и наклон на оста) и техния общ ефект върху дължината на денонощието. Приличат на синусоиди, но не са точно. ## Заключение

Математиката намира широко приложение в изучаването на небесните тела. В днешно време физиката е толкова развита, че може да опише всякакви аномалии в орбитите, които традиционните модели не успяват да обяснят. Там също се използва математика от много по-сложен характер (многомерни пространства, изкривяване на пространството в теорията на относителноста и т.н.). 

Въпреки това, дори само с гимназиални знания и наблюдения можем да извлечем много информация за близки и далечни звезди и планети.
